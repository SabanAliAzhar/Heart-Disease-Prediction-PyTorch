```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
import torch.optim as optim
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
#Explore dataframe
df=pd.read_excel("/content/heart_dataset.xlsx")
df.info()
```

```python
df.head()
```

```python
#create target vactor
df['target']=df['num'].apply(lambda x: 1 if x>0 else 0)
```

```python
#check for null values
df.isnull().sum()
df.dropna(inplace=True)
```

```python
#specify target and input features
X=df.drop(columns=['num','target','Unnamed: 0']).values
y=df['target'].values.reshape(-1,1)
```

```python
scaler=StandardScaler()
X_transformed=scaler.fit_transform(X)
```

```python
#convert to tensors
X_tensor=torch.tensor(X_transformed,dtype=torch.float32)
y_tensor=torch.tensor(y,dtype=torch.float32)
```

```python
#train, test split
X_train,X_test,y_train,y_test=train_test_split(X_tensor,y_tensor,test_size=0.2,random_state=42)
print("Train size:", X_train.shape[0])
print("Test size:", X_test.shape[0])
```

```python
#define the model
model=nn.Sequential(
    nn.Linear(13,16),
    nn.ReLU(),
    nn.Linear(16,1),
    nn.Sigmoid()
)
print(model)
```

```python
#use Adam Optimizer
optimizer=optim.Adam(model.parameters(),lr=0.01)
loss=nn.BCELoss()
```

```python
#do training
no_epochs=300
train_loss=[]
test_loss=[]
for epoch in range(no_epochs):
  model.train()
  y_pred=model(X_train)
  optimizer.zero_grad()
  loss_value=loss(y_pred,y_train)
  loss_value.backward()
  optimizer.step()

  model.eval()
  with torch.no_grad():
    y_test_pred=model(X_test)
    test_loss_value=loss(y_test_pred,y_test)

  test_loss.append(test_loss_value.item())

  train_loss.append(loss_value.item())
  if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{no_epochs}], Loss: {loss_value.item():.4f}")


y_test_prob=(y_test_pred>0.5).float()
```

```python
#Plot train loss
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(range(1, no_epochs + 1), train_loss, label="Train Loss")

plt.title("Training Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Binary Cross Entropy Loss")
plt.grid(True)
plt.legend()
plt.show()
```

```python
confusion_matrix(y_test, y_test_prob)
```

```python
#plot confusion matrix
matrix=confusion_matrix(y_test, y_test_prob)
disp=ConfusionMatrixDisplay(confusion_matrix=matrix, display_labels=["No Disease", "Disease"])
disp.plot(cmap="Blues")
plt.title("Confusion Matrix on test Set")
plt.grid(False)
plt.show()
```

```python
#check for test accuracy
test_accuracy=accuracy_score(y_test,y_test_prob)
print(f"Test Accuracy: {test_accuracy}")
```
